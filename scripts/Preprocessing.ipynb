# Initializing

### Load packages

```python
from pandas import Series, DataFrame
import pandas as pd
import time
import datetime
```

### Load data

```python
skillHeader = ['author',
 'category',
 'commands',
 'description',
 'id',
 'invocation',
 'rating',
 'ratingPercentage_1',
 'ratingPercentage_2',
 'ratingPercentage_3',
 'ratingPercentage_4',
 'ratingPercentage_5',
 'reviewNum',
 'timestamp',
 'title']
```

```python
reviewHeader = ['author',
 'commentNum',
 'content',
 'date',
 'id',
 'rating',
 'skill_title',
 'timestamp',
 'title',
 'vote']
```

```python
commentHeader = ['author', 'content', 'date', 'id', 'review_title']
```

```python
skills_1 = pd.read_csv(filepath_or_buffer='../datasets/raw/v01/skills_v01_df.csv',sep=',',encoding='utf-8',header = 0, names = skillHeader)
reviews_1 = pd.read_csv(filepath_or_buffer='../datasets/raw/v01/reviews_v01_df.csv',sep=',',encoding='utf-8',header = 0, names = reviewHeader)
comments_1 = pd.read_csv(filepath_or_buffer='../datasets/raw/v01/comments_v01_df.csv',sep=',',encoding='utf-8',header = 0, names = commentHeader, index_col= False)
```

```python
skills_2 = pd.read_csv(filepath_or_buffer='../datasets/raw/v01/skills_v01_2_df.csv',sep=',',encoding='utf-8',header = 0, names = skillHeader)
reviews_2 = pd.read_csv(filepath_or_buffer='../datasets/raw/v01/reviews_v01_2_df.csv',sep=',',encoding='utf-8',header = 0, names = reviewHeader)
comments_2 = pd.read_csv(filepath_or_buffer='../datasets/raw/v01/comments_v01_2_df.csv',sep=',',encoding='utf-8',header = 0, names = commentHeader, index_col= False)
```

```python
skills_3 = pd.read_csv(filepath_or_buffer='../datasets/raw/v01/skills_raw_Sleep and Relaxation Sounds.csv',sep=',',encoding='utf-8',header = 0, names = skillHeader)
reviews_3 = pd.read_csv(filepath_or_buffer='../datasets/raw/v01/reviews_raw_Sleep and Relaxation Sounds.csv',sep=',',encoding='utf-8',header = 0, names = reviewHeader)
comments_3 = pd.read_csv(filepath_or_buffer='../datasets/raw/v01/comments_raw_Sleep and Relaxation Sounds.csv',sep=',',encoding='utf-8',header = 0, names = commentHeader, index_col= False)
```

### Data Concatenate

```python
skills = skills_1.append(skills_2.append(skills_3, ignore_index=True), ignore_index=True)
reviews = reviews_1.append(reviews_2.append(reviews_3, ignore_index=True), ignore_index=True)
comments = comments_1.append(comments_2.append(comments_3, ignore_index=True), ignore_index=True)
```

### Data Preprocessing

#### author

```python
skills.ix[:,'author'].str.replace('by ', '')
```

```python
skills.ix[:,'author'] 
```

```python
skills.ix[:,'author'] = skills.ix[:,'author'].str.replace('by ', '')
```

#### Date

```python
datetime.datetime.strptime('On January 10, 2017', "On %B %d, %Y")
```

```python
datetime.datetime.strptime('On January 10, 17', "On %B %d, %y").strftime('%Y-%m-%d')
```

```python
datetime.datetime.strptime(str(reviews.ix[0,'date']), "On %B %d, %Y")
```

```python
reviews.ix[:,'date'] = reviews.ix[:,'date'].apply(lambda x: datetime.datetime.strptime(x, "On %B %d, %Y"))
```

#### Rating

```python
skills.ix[:,'rating'] = skills.ix[:,'rating'].str.replace(' out of 5 stars', '')
skills.ix[:,'rating'] = pd.to_numeric(skills.ix[:,'rating'])
```

```python
reviews.ix[:,'rating'] = reviews.ix[:,'rating'].str.replace(' out of 5 stars', '')
reviews.ix[:,'rating'] = pd.to_numeric(reviews.ix[:,'rating'])
```

#### Rating Percentage

```python
skills.ix[:,'ratingPercentage_1'] = skills.ix[:,'ratingPercentage_1'].str.replace('%', '')
skills.ix[:,'ratingPercentage_1'] = pd.to_numeric(skills.ix[:,'ratingPercentage_1'])
skills.ix[:,'ratingPercentage_2'] = skills.ix[:,'ratingPercentage_2'].str.replace('%', '')
skills.ix[:,'ratingPercentage_2'] = pd.to_numeric(skills.ix[:,'ratingPercentage_2'])
skills.ix[:,'ratingPercentage_3'] = skills.ix[:,'ratingPercentage_3'].str.replace('%', '')
skills.ix[:,'ratingPercentage_3'] = pd.to_numeric(skills.ix[:,'ratingPercentage_3'])
skills.ix[:,'ratingPercentage_4'] = skills.ix[:,'ratingPercentage_4'].str.replace('%', '')
skills.ix[:,'ratingPercentage_4'] = pd.to_numeric(skills.ix[:,'ratingPercentage_4'])
skills.ix[:,'ratingPercentage_5'] = skills.ix[:,'ratingPercentage_5'].str.replace('%', '')
skills.ix[:,'ratingPercentage_5'] = pd.to_numeric(skills.ix[:,'ratingPercentage_5'])
```

#### CommentNum

```python
reviews.ix[:,'commentNum']
```

```python
reviews.ix[:,'commentNum'] = pd.to_numeric(reviews.ix[:,'commentNum'])
```

#### Vote

```python
reviews.ix[:,'vote'] = reviews.ix[:,'vote'].str.replace(' people found this helpful.', '')
reviews.ix[:,'vote'] = reviews.ix[:,'vote'].str.replace('One person found this helpful.', '1')
# reviews.ix[:,'vote'] = reviews.ix[:,'vote'].str.replace('NaN', '0')
reviews.ix[:,'vote'] = pd.to_numeric(reviews.ix[:,'vote'])
```

#### timestamp

```python
# skills.ix[:,'timestamp']
skills.ix[:,'timestamp'] = pd.to_numeric(skills.ix[:,'timestamp'])
```

```python
# reviews.ix[:,'timestamp']
reviews.ix[:,'timestamp'] = pd.to_numeric(reviews.ix[:,'timestamp'])
```
